```python
import matplotlib.pyplot as plt
import numpy as np
```

### Graficar X Y

```python
x = [1, 2, 3, 4]
y = [2, 4, 6 ,8]
```

```python
plt.plot(x, y)
plt.show() # mostrar Gráfico
```

### Agregar Etiquetas - Título - Grilla

```python
plt.plot(x, y)
plt.xlabel('Etiqueta X') # x label
plt.ylabel('Etiqueta Y') # y label
plt.title('Título', {'fontsize': 18.5}) # title, fontsize=tamaño del título
plt.grid(b=True)
plt.show()
```

### Estílos y gráficos en una sola figura

```python
x = np.linspace(0, 2, 100)
plt.plot(x, x, 'r', label='linear')
plt.plot(x, x**2, 'y', label='cuadrática')
plt.plot(x, x**3, 'g', label='cúbica')

plt.xlabel('X')
plt.ylabel('Y')

plt.title("Simple Plot")

plt.legend(loc=7, )

plt.show()
```

```python
x = np.linspace(0, 2, 10)
plt.plot(x, x, 'or')
plt.plot(x, x**2, '4y', markersize=12.5) # markersize => tamaño del marcador
plt.plot(x, x**3, '*-g')

plt.xlabel('X')
plt.ylabel('Y')

plt.title("Simple Plot")

plt.legend(['linear', 'cuadrática', 'cúbica'])

plt.grid(b=True)

plt.show()
```

### Estítlos y graficos en varias figuras

```python
x = np.linspace(0, 2, 10)
plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.plot(x, x, 'or')

plt.subplot(132)
plt.plot(x, x**2, '4y', markersize=12.5)

plt.subplot(133)
plt.plot(x, x**3, '*-g')

plt.suptitle('Título General')
plt.show()
```
